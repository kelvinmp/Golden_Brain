# Apertura del Dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Data.csv')
data.set_index('PisoApartamento', inplace=True)
data
```

# Medias por apartamento

```python
# Media de cada valor (Estatura, peso, edades).
estatura_media = data.groupby('PisoApartamento')['Estatura(m)'].mean()
pesos_media = data.groupby('PisoApartamento')['Peso(kg)'].mean()
edades_media = data.groupby('PisoApartamento')['Edad'].mean()

# Graficos de barras.
plt.figure(figsize=(50, 15))
plt.subplot(331)
plt.bar(pesos_media.index, pesos_media.values, color = 'orange')
plt.xticks(size= 7, rotation= 50)
plt.title('Estadisticas de Media de Pesos')


plt.subplot(334)
plt.bar(estatura_media.index, estatura_media.values)
plt.xticks(size= 7, rotation= 50)
plt.title('Estadisticas de Media de Estatura')

plt.subplot(337)
plt.bar(edades_media.index, edades_media.values, color= 'green')
plt.xticks(size= 7, rotation= 50)
plt.title('Estadisticas de Media de Edades')

plt.show()
```

# Graficos de dispersiÃ³n (edades, altura, peso)

```python

indices_nounique = data.index.tolist() # Lista de indices completa

# Datos completos
pesos_completos = data['Peso(kg)'].tolist()
estaturas_completas = data['Estatura(m)'].tolist()
edades_completas = data['Edad'].tolist()

plt.figure(figsize=(17.5, 4.5)) # TamaÃ±o de la figura

# CreaciÃ³n de los graficos dispersos detallados
plt.subplot(131)
plt.scatter(indices_nounique, pesos_completos, color = 'orange')
plt.title('Pesos', color = 'orange')
plt.xticks(size= 6, rotation= 50)

plt.subplot(132)
plt.scatter(indices_nounique, edades_completas, color = 'green')
plt.title('Edades', color= 'green')
plt.xticks(size= 6, rotation= 50)

plt.subplot(133)
plt.scatter(indices_nounique, estaturas_completas)
plt.title('Estatura', color= 'blue')
plt.xticks(size= 6, rotation= 50)

plt.suptitle('Estadisticas detalladas', color= 'red')

plt.show()
```

```python
indices = data.index.unique().tolist() # Lista de indices unicas

plt.figure(figsize=(17.5, 4.5))

# CreaciÃ³n de los graficos dispersos medias
plt.subplot(131)
plt.scatter(indices, pesos_media.values, color = 'orange')
plt.title('Pesos', color = 'orange')
plt.xticks(size= 6, rotation= 50)

plt.subplot(132)
plt.scatter(indices, edades_media.values, color = 'green')
plt.title('Edades', color= 'green')
plt.xticks(size= 6, rotation= 50)

plt.subplot(133)
plt.scatter(indices, estatura_media.values)
plt.title('Estatura', color= 'blue')
plt.xticks(size= 6, rotation= 50)

plt.suptitle('Estadisticas medias', color= 'red')

plt.show()
```

# Estadisticas detalladas con grafico de caja

```python
plt.figure(figsize=(17.5, 4.5))

# Creacion de estadisticas detalladas con grafico de caja
plt.subplot(131)
plt.boxplot(pesos_completos, labels = ['Apartamentos'], flierprops=dict(markerfacecolor='red'))
plt.ylabel('Pesos', color = 'orange')


plt.subplot(132)
plt.boxplot(edades_completas, labels = ['Apartamentos'], flierprops=dict(markerfacecolor='red'))
plt.ylabel('Edades', color= 'green')

plt.subplot(133)
plt.boxplot(estaturas_completas, labels = ['Apartamentos'], flierprops=dict(markerfacecolor='red'))
plt.ylabel('Estatura', color= 'blue')

plt.supylabels('Estadisticas detalladas', color= 'red')

plt.show()
```
