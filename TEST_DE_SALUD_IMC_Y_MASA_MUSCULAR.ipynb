```python
ACTIVIDAD_SEDENTARIA = 1.2
ACTIVIDAD_MODERADA = 1.55
ACTIVIDAD_ALTA = 1.9

def calcular_imc(peso, altura):
    imc = peso / (altura ** 2)
    return imc

def calcular_calorias(peso, altura, edad, sexo, actividad_fisica, objetivo):
    if sexo == 'M':
        bmr = 88.362 + (13.397 * peso) + (4.799 * altura * 100) - (5.677 * edad)
    elif sexo == 'F':
        bmr = 447.593 + (9.247 * peso) + (3.098 * altura * 100) - (4.330 * edad)
    else:
        return "Sexo inválido"

    if actividad_fisica == 'sedentario':
        calorias = bmr * 1.2
    elif actividad_fisica == 'moderado':
        calorias = bmr * 1.55
    else:
        calorias = bmr * 1.9

    if objetivo == 'subir':
        calorias += 500
    elif objetivo == 'bajar':
        calorias -= 500

    return calorias

def calcular_proteinas(peso, nivel_actividad):
    if nivel_actividad == 'alto':
        return peso * 2.2
    elif nivel_actividad == 'moderado':
        return peso * 1.8
    else:
        return peso * 1.6

def consejos_imc(imc):
    if imc < 18.5:
        return "Su IMC indica que está por debajo del peso saludable. Es recomendable aumentar la ingesta calórica y consultar con un profesional de la salud."
    elif 18.5 <= imc < 24.9:
        return "Su IMC es saludable. Mantenga un estilo de vida equilibrado con una dieta balanceada y ejercicio regular."
    elif 25 <= imc < 29.9:
        return "Su IMC indica sobrepeso. Es recomendable consultar con un nutricionista y hacer cambios en su dieta y rutina de ejercicios."
    else:
        return "Su IMC indica obesidad. Es importante buscar orientación médica para desarrollar un plan adecuado para perder peso de manera saludable."

def consejos_entrenador(objetivo):
    if objetivo == 'subir':
        return "Como dijo Arnold Schwarzenegger, 'El dolor es temporal, el orgullo es para siempre.' Mantén el enfoque y la consistencia en tu dieta y entrenamiento."
    elif objetivo == 'bajar':
        return "Recuerda las palabras del Dr. William H. Masters: 'El control de tu peso está en tus manos, pero la clave es la paciencia y la persistencia.'"
    elif objetivo == 'mantener':
        return "Según la Dra. Jane Fonda, 'La salud es la mejor inversión que puedes hacer.' Mantén un balance entre dieta y ejercicio para lograr bienestar a largo plazo."

def consejos_padecimiento(padecimiento):
    if padecimiento == 'hipertension':
        return "Con hipertensión, es importante controlar la ingesta de sal y realizar ejercicio moderado. Consulte con un médico para un plan adecuado."
    elif padecimiento == 'cervical':
        return "Si tiene problemas cervicales, evite ejercicios que puedan generar tensión en el cuello y consulte con un fisioterapeuta para ejercicios específicos."
    elif padecimiento == 'corazon':
        return "Si tiene problemas de corazón, es crucial evitar ejercicios intensos. Consulte con un cardiólogo antes de empezar cualquier rutina de ejercicios."
    elif padecimiento == 'ninguno':
        return "Es recomendable llevar un estilo de vida saludable, realizar ejercicio regular y mantener una dieta equilibrada."
    elif padecimiento == 'otro':
        return "Es recomendable consultar con un especialista para desarrollar un plan adecuado para su padecimiento."

def receta_objetivo(objetivo):
    recetas = {
        'subir': {
            'nombre': 'Batido Calórico de Plátano y Avena',
            'ingredientes': ['1 plátano', '1 taza de avena', '1 cucharada de mantequilla de maní', '1 vaso de leche entera', 'Miel al gusto'],
            'preparacion': '1. Licúa todos los ingredientes hasta obtener una mezcla suave. 2. Sirve en un vaso y disfruta como merienda o desayuno.'
        },
        'bajar': {
            'nombre': 'Ensalada de Pollo y Aguacate',
            'ingredientes': ['150g de pechuga de pollo', '1 aguacate', 'Espinacas frescas', 'Tomates cherry', 'Aceite de oliva', 'Sal y pimienta al gusto'],
            'preparacion': '1. Cocina la pechuga de pollo a la plancha. 2. Mezcla todos los ingredientes en un bol y adereza con aceite de oliva, sal y pimienta.'
        },
        'mantener': {
            'nombre': 'Tostadas de Aguacate y Huevo',
            'ingredientes': ['1 aguacate', '2 rebanadas de pan integral', '2 huevos', 'Aceite de oliva', 'Sal y pimienta al gusto'],
            'preparacion': '1. Tuesta las rebanadas de pan integral. 2. En una sartén, cocina los huevos al gusto. 3. Unta el aguacate sobre las tostadas y coloca los huevos encima.'
        }
    }

    return recetas.get(objetivo, None)

def consejos_expertos():
    return [
        "La Organización Mundial de la Salud (OMS) recomienda que los adultos realicen al menos 150 minutos de actividad física moderada o 75 minutos de actividad intensa cada semana.",
        "Según el Colegio Americano de Medicina Deportiva, se debe equilibrar el entrenamiento de fuerza con el ejercicio cardiovascular para mantener un corazón saludable y mejorar la composición corporal.",
        "La Asociación Americana del Corazón sugiere que una dieta baja en grasas saturadas, sodio y azúcares añadidos es clave para reducir el riesgo de enfermedades cardiovasculares.",
        "Según la Clínica Mayo, una hidratación adecuada es fundamental para el rendimiento físico, la digestión y la salud general. Se recomienda consumir al menos 8 vasos de agua al día, o más según la actividad física y el clima.",
        "El Dr. Michael Greger, autor de 'Cómo no morir', recomienda una dieta rica en alimentos de origen vegetal, que puede ayudar a prevenir muchas enfermedades crónicas.",
        "La Asociación Nacional de Ejercicio y Salud (NAHS) recomienda incluir estiramientos antes y después del ejercicio para prevenir lesiones y mejorar la flexibilidad.",
        "El Instituto Nacional del Corazón, los Pulmones y la Sangre recomienda reducir la ingesta de azúcares añadidos, ya que están directamente relacionados con el aumento de peso y enfermedades como la diabetes tipo 2.",
        "El nutricionista David Katz sugiere que comer cinco porciones de frutas y verduras al día puede mejorar el bienestar general y reducir el riesgo de enfermedades crónicas.",
        "La Sociedad Internacional de Nutrición Deportiva enfatiza que las proteínas son esenciales para la reparación y crecimiento muscular, especialmente después de un entrenamiento intenso.",
        "Según la Academia Americana de Nutrición y Dietética, consumir una variedad de alimentos de todos los grupos alimenticios ayuda a garantizar una ingesta adecuada de nutrientes."
    ]

def calculadora():
    print("Bienvenido a la calculadora de IMC y masa muscular")

    while True:
        try:
            peso = float(input("Ingrese su peso en kg: "))
            altura = float(input("Ingrese su altura en metros: "))
            break
        except ValueError:
            print("Entrada inválida. Por favor ingrese valores numéricos para peso y altura.")

    imc = calcular_imc(peso, altura)
    print(f"Su IMC es: {imc:.2f}")

    while True:
        objetivo = input("¿Cuál es su objetivo? (subir/mantener/bajar): ").lower()
        if objetivo.startswith(('subir', 'mantener', 'bajar')):
            break
        else:
            print("Mensaje inválido. Por favor ingrese 'subir', 'mantener' o 'bajar'.")

    receta = receta_objetivo(objetivo)
    if receta:
        print(f"\nReceta recomendada para {objetivo} de peso:")
        print(f"Nombre: {receta['nombre']}")
        print("Ingredientes: ", ', '.join(receta['ingredientes']))
        print("Preparación: ", receta['preparacion'])

    print("\nConsejos según su IMC:")
    print(consejos_imc(imc))

    while True:
        padecimiento = input("¿Tiene algún padecimiento? (hipertension/cervical/corazon/ninguno/otro): ").lower()
        if padecimiento.startswith(('hipertension', 'cervical', 'corazon', 'ninguno', 'otro')):
            break
        else:
            print("Mensaje inválido. Por favor ingrese un padecimiento válido.")

    print(consejos_padecimiento(padecimiento))
    print("\nConsejos del entrenador:")
    print(consejos_entrenador(objetivo))

    desea_consejos = input("\n¿Quieres consejos adicionales? (si/no): ").lower()
    if desea_consejos == 'si':
        for consejo in consejos_expertos():
            print(f"- {consejo}")

# Llamada principal a la función calculadora
calculadora()
```
